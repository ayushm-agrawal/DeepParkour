```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.signal import lfilter
from baselines.common import plot_util as pu
```

```python
log_file_1 = 'logs/5_15_agent/progress.csv'
log_file_2 = 'logs/10_15_agent/progress.csv'
log_file_3 = 'logs/15_15_agent/progress.csv'
# log_file_1 = 'logs/5_1_agent/progress.csv'
# log_file_2 = 'logs/10_1_agent/progress.csv'
# log_file_3 = 'logs/15_1_agent/progress.csv'
data_1 = pd.read_csv(log_file_1)
data_2 = pd.read_csv(log_file_2)
data_3 = pd.read_csv(log_file_3)
```

```python
n=10
b=[1.0 /n] *n
a = 1
plt.figure(1, figsize=(15, 4))
yy = lfilter(b,a,data_1['EpRewMean']*10)
plt.subplot(131)
plt.plot(data_1['TimestepsSoFar'], data_1['EpRewMean']*10)
plt.xlabel('Timesteps')
plt.ylabel('Episode Reward Mean')
yy = lfilter(b,a,data_2['EpRewMean']*10)
plt.subplot(132)
plt.plot(data_2['TimestepsSoFar'], data_2['EpRewMean']*10, 'r')
plt.xlabel('Timesteps')
yy = lfilter(b,a,data_3['EpRewMean']*10)
plt.subplot(133)
plt.plot(data_3['TimestepsSoFar'], data_3['EpRewMean']*10,'g')
plt.xlabel('Timesteps')
# plt.savefig('time-v-reward(1).png')
```
